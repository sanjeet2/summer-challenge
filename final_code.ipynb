```python
import cv2
from ultralytics import YOLO
import matplotlib.pyplot as plt

# Load the YOLO model
model = YOLO('best_test_v8s.pt')

# Define the question regions
question_regions = [
    (450, 20, 200, 50),  # Question 1 region
    (450, 75, 200, 45),  # Question 2 region
    (450, 120, 200, 45),  # Question 3 region
    (450, 169, 200, 30),  # Question 4 region
    (450, 200, 200, 45),  # Question 5 region
    (450, 250, 200, 55),  # Question 6 region
    (450, 300, 200, 53),  # Question 7 region
    (400, 353, 230, 30),  # Question 8 region
    (450, 385, 200, 63),  # Question 9 region
    (400, 448, 230, 50)   # Question 10 region
]

image_path = r"C:\Users\sanje\Desktop\final_test_handwritten\train\images\20240328_151033.jpg"
img = cv2.imread(image_path)
img = cv2.resize(img, (640, 640))

# Extract the predicted labels for each question region
class_names = {0: 'TRUE', 1: 'FALSE', 2: 'BLANK', 3: 'ERASE'}
predicted_labels = []
correct_answers = ['TRUE', 'FALSE', 'TRUE', 'FALSE', 'TRUE', 'FALSE', 'TRUE', 'FALSE', 'TRUE', 'FALSE']  # Replace with the correct answers

total_questions = len(question_regions)
correct_count = 0

for x, y, w, h in question_regions:
    roi = img[y:y+h, x:x+w]
    results = model(roi)[0]
    labels = [class_names[int(detection[5])] for detection in results.boxes.data]
    
    if 'ERASE' in labels and ('TRUE' in labels or 'FALSE' in labels):
        labels.remove('ERASE')  # Remove 'ERASE' if 'TRUE' or 'FALSE' is also detected
    
    if 'TRUE' in labels and 'FALSE' in labels:
        # If both 'TRUE' and 'FALSE' are detected, consider the one with higher confidence
        true_conf = next((detection.conf for detection in results.boxes.data if class_names[int(detection[5])] == 'TRUE'), 0)
        false_conf = next((detection.conf for detection in results.boxes.data if class_names[int(detection[5])] == 'FALSE'), 0)
        if true_conf > false_conf:
            labels = ['TRUE']
        else:
            labels = ['FALSE']
    
    if not labels:
        labels = ['ERASE']  # Consider undetected labels as 'ERASE'
    
    predicted_labels.append(labels)

    # Check if the predicted label matches the correct answer
    if labels == [correct_answers[len(predicted_labels) - 1]]:
        correct_count += 1

# Draw the question regions on the original image
for x, y, w, h in question_regions:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)

# Display the image with the question regions
plt.figure(figsize=(10, 8))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()

# Print the predicted labels for each question
for i, labels in enumerate(predicted_labels):
    print(f"Question {i+1}: {', '.join(labels)}")

# Calculate the total marks
total_marks = (correct_count / total_questions) * 100
print(f"Total Marks: {total_marks:.2f}%")
```

```python
from ultralytics import YOLO
import cv2
import cvzone 
import math





image_path = r"C:\Users\sanje\Desktop\final_test_handwritten\train\images\20240328_151033.jpg"
img = cv2.imread(image_path)

# Load YOLO model
model = YOLO('best_test_v8s.pt')


class_names = {0: 'TRUE' , 1: 'FALSE' , 2: 'BLANK' }


results = model(img, stream=True)


for r in results:
    boxes = r.boxes
    for box in boxes:
        x1, y1, x2, y2 = box.xyxy[0]
        x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
        w, h = x2 - x1, y2 - y1
        cvzone.cornerRect(img, (x1, y1, w, h))

        # Show confidence value
        conf = (math.ceil(box.conf[0] * 100)) / 100
        print(conf)

        # Show object names
        cls = int(box.cls[0])
        cvzone.putTextRect(img, f'{class_names[cls]} {conf}', (max(0, x1), max(30, y1)), scale=1, thickness=2)

# Display the image

img_resized = cv2.resize(img,(640,640))
plt.imshow(img_resized)
# cv2.imshow("Detected Image", img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```
